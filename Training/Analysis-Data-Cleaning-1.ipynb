```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
import nltk
import webbrowser
import os

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
apps_df = pd.read_csv("Datasets/Play Store Data.csv")
review_df = pd.read_csv("Datasets/User Reviews.csv")
```

```python
apps_df.dropna(subset=["Rating"])
```

```python
for col in apps_df.columns:
    apps_df[col].fillna(apps_df[col].mode()[0], inplace=True)

apps_df.drop_duplicates(inplace=True)

apps_df=apps_df[apps_df['Rating'] <= 5]
```

```python
review_df.dropna(subset=["Translated_Review"], inplace=True)
```

```python
apps_df.dtypes
```

```python
apps_df['Installs'] = apps_df['Installs'].str.replace(',', '').str.replace('+', '').astype(int)
```

```python
apps_df['Price'] = apps_df['Price'].str.replace('$', '').astype(float)
```

```python
apps_df.dtypes
```

```python
merged_df = pd.merge(apps_df, review_df, on="App", how="inner")
```

```python
merged_df
```

```python
temp_df = merged_df[(merged_df.Size != "Varies with device") & (~merged_df.Size.str.contains("M"))]
temp_df
```

```python
def convert_size(size):
    if "M" in size:
        return float(size.replace("M", ""))
    elif "K" in size:
        return float(size.replace("K", "")) / 1024
    else:
        return np.nan
    
apps_df.Size = apps_df.Size.apply(convert_size)
apps_df
```

```python
apps_df["Reviews"] = apps_df["Reviews"].astype(int)
```

```python
apps_df["Log Installs"] = np.log(apps_df["Installs"])
apps_df['Log Reviews'] = np.log(apps_df['Reviews'])
```

```python
def rating_group(rating):
    if rating >= 4:
        return "Top Rated"
    elif rating >= 3:
        return "Above Average"
    elif rating >= 2:
        return "Average"
    else:
        return "Bad"
    
apps_df["Rating Group"] = apps_df["Rating"].apply(rating_group)
```

```python
apps_df["Revenue"] = apps_df["Price"] * apps_df["Installs"]
```
