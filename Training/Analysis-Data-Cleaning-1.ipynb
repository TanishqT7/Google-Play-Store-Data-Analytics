```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.io as pio
import nltk
import webbrowser
import os

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
apps_df = pd.read_csv("Datasets/Play Store Data.csv")
review_df = pd.read_csv("Datasets/User Reviews.csv")
```

```python
apps_df
```

```python
review_df.head()
```

```python
apps_df.isna().sum()
```

```python
apps_df.dropna(subset=["Rating"])
```

```python
for col in apps_df.columns:
    apps_df[col].fillna(apps_df[col].mode()[0], inplace=True)

apps_df.drop_duplicates(inplace=True)

apps_df=apps_df[apps_df['Rating'] <= 5]
```

```python
review_df.dropna(subset=["Translated_Review"], inplace=True)
```

```python
apps_df.dtypes
```

```python
apps_df['Installs'] = apps_df['Installs'].str.replace(',', '').str.replace('+', '').astype(int)
```

```python
apps_df['Price'] = apps_df['Price'].str.replace('$', '').astype(float)
```

```python
apps_df.dtypes
```

```python
merged_df = pd.merge(apps_df, review_df, on="App", how="inner")
```

```python
merged_df
```

```python
temp_df = merged_df[(merged_df.Size != "Varies with device") & (~merged_df.Size.str.contains("M"))]
temp_df
```

```python
def convert_size(size):
    if "M" in size:
        return float(size.replace("M", ""))
    elif "K" in size:
        return float(size.replace("K", "")) / 1024
    else:
        return np.nan
    
apps_df.Size = apps_df.Size.apply(convert_size)
apps_df
```

```python
apps_df["Reviews"] = apps_df["Reviews"].astype(int)
```

```python
apps_df["Log Installs"] = np.log(apps_df["Installs"])
apps_df['Log Reviews'] = np.log(apps_df['Reviews'])
```

```python
def rating_group(rating):
    if rating >= 4:
        return "Top Rated"
    elif rating >= 3:
        return "Above Average"
    elif rating >= 2:
        return "Average"
    else:
        return "Bad"
    
apps_df["Rating Group"] = apps_df["Rating"].apply(rating_group)
```

```python
apps_df["Revenue"] = apps_df["Price"] * apps_df["Installs"]
```

```python
apps_df
```

# Plotly Training

```python
html_files_path = './html_files'
if not os.path.exists(html_files_path):
    os.makedirs(html_files_path)
```

```python
plot_containers = ""
```

```python
def save_plot_as_html(fig, filename, insight):
    global plot_containers

    file_path = os.path.join(html_files_path, filename)
    html_content = pio.to_html(fig, include_plotlyjs = 'inline', full_html = False)

    plot_containers += f"""
    <div class="plot-container" id="{filename}" onclick="openPlot('{file_path}');">
        <div class="plot">{html_content}</div>
        <div class="insight">{insight}</div>
    </div>
    """

    fig.write_html(file_path, full_html = False, include_plotlyjs = 'inline')
```

```python

plot_width = 800
plot_height = 600
plot_bg_color = 'black'
text_color = 'white'
title_font = {'size':16}
axis_font = {'size':12}
```

```python
# Plot 1

from pydoc import text


category_Count = apps_df["Category"].value_counts().nlargest(10)

fig1 = px.bar(
    x = category_Count.index,
    y = category_Count.values,
    title = "Top 10 Categories",
    height = plot_height,
    width = plot_width,
    template = "plotly_dark",
    labels = {"x": "Category", "y": "Number of Apps"},
    color_discrete_sequence = ["#FFA500"]
)

fig1.update_layout(
    plot_bgcolor=plot_bg_color,
    paper_bgcolor=plot_bg_color,
    font_color=text_color,
    title_font=title_font,
    xaxis=dict(title_font=axis_font),
    yaxis=dict(title_font=axis_font),
    margin=dict(l=10,r=10,t=30,b=10)
)

save_plot_as_html(fig1, "Category Graph 1.html", "The top categories on the Play Store are dominated by tools, entertainment, and productivity apps")
```

```python
# Plot 2

type_Count = apps_df["Type"].value_counts()

fig2 = px.pie(
    values = type_Count.values,
    names = type_Count.index,
    title = "App Type Distribution",
    height = plot_height,
    width = plot_width,
    color_discrete_sequence = ["#FFA500"]
)

fig2.update_layout(
    plot_bgcolor=plot_bg_color,
    paper_bgcolor=plot_bg_color,
    font_color=text_color,
    title_font=title_font,
    margin=dict(l=10,r=10,t=30,b=10)
)

save_plot_as_html(fig2, "Type Graph 1.html", "Most apps on playstore are free, indicating a strategy to attract users first and monetize through ads or in app purchases.")
```
